```python
import folium
import geopandas as gpd
import pandas as pd
from folium import GeoJson
from shapely.geometry import point
```

```python
radius = 1000
```

```python
#------------MAPA
fmap = folium.Map(location=[-23.5, -46.6], zoom_start=10 )
```

```python
#------------COROPLÉTICO
caminho_arquivo_shapefile = 'demographics/SIRGAS_SHP_densidade_demografica_2010.shp'

dados_shapefile_raw = gpd.read_file(caminho_arquivo_shapefile)

#DataFrame de testes, para execução completa, comente as 3 linhas e altere dados_shapefile_raw para dados_shapefile
dados_shapefile = gpd.GeoDataFrame()
dados_shapefile = dados_shapefile_raw[dados_shapefile_raw.index % 3 == 0]
dados_shapefile = dados_shapefile.reset_index(drop=True)

dados_shapefile.crs = {'init': 'epsg:31983'}#isso eu creio que seja problema com o GeoPandas, mais especificamente
                                            #com a maneira que ele chama o método CRS da pyproj 
dados_shapefile.to_crs(epsg='4326', inplace=True)

folium.Choropleth(
    geo_data=dados_shapefile,
    name='choropleth',
    data=dados_shapefile,
    columns=['id', 'populacao'],
    key_on='feature.properties.id',
    fill_color='OrRd',
    fill_opacity=0.9,
    line_opacity=0.1,
    legend_name='Densidade Populacional (hab/ha)'
).add_to(fmap)
```

```python
#------------PONTOS

path_subway_stations = 'interest_points/SIRGAS_SHP_estacaometro_point.shp'
path_train_stations = 'interest_points/SIRGAS_SHP_estacaotrem.shp'
path_bus_terminals = 'interest_points/sirgas_terminal_onibus.shp'

file_paths = [path_bus_terminals, path_subway_stations, path_train_stations]

for shapefile_path in file_paths:
    gdf = gpd.read_file(shapefile_path)
    gdf.crs = {'init': 'epsg:31983'}  
    gdf.to_crs(epsg='4326', inplace=True)
    gjson = GeoJson(gdf)   
    gjson.add_to(fmap)
    for index, row in gdf.iterrows():
        latitude = row['geometry'].x
        longitude = row['geometry'].y
        circle = folium.Circle(
            location=[longitude, latitude],
            radius=radius,
            color='blue',  
            fill=True,   
            fill_color='blue',  
            fill_opacity=0.2   
        )
        circle.add_to(fmap)
```

```python
radius = 1000

for shapfile_path in file_paths:
    points_gdf = gpd.read_file(shapefile_path)

    points_gdf.crs = {'init': 'epsg:31983'}
    points_gdf.to_crs(epsg='32634', inplace=True)

    buffers_gdf = gpd.GeoDataFrame(geometry=[])

    buffers_gdf.crs = {'init': 'epsg:31983'}
    buffers_gdf.to_crs(epsg='32634', inplace=True)

    for index, row in points_gdf.iterrows():
        buffer_geometry = row['geometry'].buffer(radius)
        buffers_gdf = pd.concat([buffers_gdf, gpd.GeoDataFrame(geometry=[buffer_geometry])], ignore_index=True)

# Sistema de coordenadas padrão do folium
buffers_gdf.to_crs(epsg='4326', inplace=True)
display(buffers_gdf)
```

```python

intersecting_gdf = gpd.GeoDataFrame(columns=dados_shapefile.columns)

pcenter_points = []
ccenter_points = []

for index, polygon in buffers_gdf.iterrows():
    intersects = dados_shapefile['geometry'].intersects(polygon['geometry'])
    
    if intersects.any():    
        intersection_list = dados_shapefile[intersects]
        densest_sector = intersection_list.loc[intersection_list['habit_hect'].idxmax()]
        intersecting_gdf = pd.concat([intersecting_gdf, densest_sector.to_frame().T], ignore_index=True)
        
        pcenter = densest_sector['geometry'].centroid
        pcenter_points.append(pcenter)
        
        ccenter = polygon['geometry'].centroid
        ccenter_points.append(ccenter)
        
intersecting_gdf['circle_center'] = ccenter_points
intersecting_gdf['polygons_center'] = pcenter_points
display(intersecting_gdf)
```
