```python
#1:n most relevant point to closest points
#In this code, the DataFrame should be ordered by the relevant field descending order


import geopandas as gpd
import json

caminho_arquivo_shapefile = 'demographics/SIRGAS_SHP_densidade_demografica_2010.shp'

polygons1_raw = gpd.read_file(caminho_arquivo_shapefile)

#DataFrame de testes, para execução completa, comente as 3 linhas e altere dados_shapefile_raw para dados_shapefile
polygons1 = gpd.GeoDataFrame()
polygons1 = polygons1_raw[polygons1_raw.index % 3 == 0]
polygons1 = polygons1.dropna()
polygons1 = polygons1.reset_index(drop=True)
polygons1.crs = {'init': 'epsg:31983'}

polygons2 = gpd.read_file("POIs.json", driver = 'GeoJSON')
polygons2.to_crs(epsg='31983', inplace=True)

result_df = gpd.GeoDataFrame(columns=['polygon1_centroid', 'polygon2_centroid'])

for index1, polygon1 in polygons1.iterrows():
    intersecting_polygons2 = polygons2[polygons2.intersects(polygon1['geometry'])]

    for index2, polygon2 in intersecting_polygons2.iterrows():
        centroid1 = polygon1['geometry'].centroid
        centroid2 = polygon2['geometry'].centroid

        result_df = result_df.append({
            'polygon1_centroid': centroid1,
            'polygon2_centroid': centroid2
        }, ignore_index=True)

result_gdf1 = gpd.GeoDataFrame(geometry=result_df['polygon1_centroid'])
result_gdf2 = gpd.GeoDataFrame(geometry=result_df['polygon2_centroid'])
```
