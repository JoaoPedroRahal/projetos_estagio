```python
import folium
import pandas as pd
import json
from unidecode import unidecode
```

```python
# Create the world map
mp = folium.Map(location=[-15, -50], zoom_start=3.5 )

# Display the new map
mp
```

```python
# Extract the cities' coordinates and names 
with open('br.json', encoding = 'utf-8') as file:
    city_data = json.load(file)

city_df = pd.DataFrame(city_data)

# Remove special characters from the names
city_df['city'] = city_df['city'].apply(lambda x: unidecode(x))

# DataFrame for connecting only capitals
capital_df = city_df[city_df['capital'] == 'admin']
capital_df = capital_df.reset_index()

#display(capital_df)
```

```python
# Create a FeatureGroup to hold the polylines
line_layer = folium.FeatureGroup(name='Polylines')

# Iterate over the cities and connect the coordinates with polylines
for i in range(len(capital_df)):
    city_A = capital_df.loc[i, 'city']
    city_B = capital_df.loc[(i + 1) % len(capital_df), 'city']
    coor_A = (float(capital_df.loc[i, 'lat']), float(capital_df.loc[i, 'lng']))
    coor_B = (float(capital_df.loc[(i + 1) % len(capital_df), 'lat']), float(capital_df.loc[(i + 1) % len(capital_df), 'lng']))
    line = folium.PolyLine([coor_A, coor_B], color='blue')
    line.add_to(line_layer)

# Add the FeatureGroup to the map
line_layer.add_to(mp)

mp
```

```python
# Save the map
mp.save('brazil_lines.html')
```
